```python
# filling missing values
df['retail_price']=df['retail_price'].fillna(df['retail_price'].median())
df['discounted_price']=df['discounted_price'].fillna(0)
df['brand'].fillna('unknown',inplace=True)
df['product_specifications'].fillna('unknown',inplace=True)
df['description'].fillna('unknown',inplace=True)
df['image'].fillna('Unknown',inplace=True)
df['product_rating'].fillna(0,inplace=True)
df['product_rating']=df['product_rating'].astype(int)
```

```python
# Top product category (count)
plt.figure(figsize=(12,8))
df['product_category_tree'].value_counts().sort_values(ascending=False).head(20).plot(kind='barh')
plt.title('Top Product Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Category wise avg Price
cat_price = (
    df.groupby('product_category_tree')['retail_price']
    .mean()
    .sort_values(ascending=False)
    .head(10)
)

plt.figure(figsize=(12,8))
cat_price.plot(kind='barh')
plt.title('Category-wise Average Price')
plt.xlabel('Average Price')
plt.show()
```
